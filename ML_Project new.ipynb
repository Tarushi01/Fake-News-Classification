```python
import pandas as pd
import numpy as np
```

```python
fake = pd.read_csv('Fake.csv')
true = pd.read_csv('True.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
# inserting label column as target feature
fake['label'] = 0
true['label'] = 1
```

```python
true.head()
```

```python
# concatenating both tables row-wise (axis=0)
news = pd.concat([fake, true], axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
# checking for null values
news.isnull().sum()
```

```python
# feature selection: we dont require title, subject and date columns so we drop them (column-wise => axis=1)
news = news.drop(['title', 'subject', 'date'], axis = 1)
```

```python
news.head()
```

```python
# since dataset is looking more organized, i.e. fake news seems to be at the top and true news is at bottom; we will reshuffle the dataset
news = news.sample(frac=1) # reshuffling
```

```python
news.head()
```

```python
# after resfuffling, the first column is jumbled, so we will reset the index values (i.e. we append a new column for numbering)
news.reset_index(inplace = True)
```

```python
news.head()
```

```python
# we drop the jumbled index column
news.drop(['index'], axis = 1, inplace = True)
```

```python
news.head()
```

```python
# Feature Extraction
```

```python
# we have a text column with us, but in ml algorithms it is better to work with numeric data.
# so we convert them into numerical vectors (tfidf).
# we do this because some words are important in the text column, the model msut be able to get the meaning out of the text
# hence there are some meaning in the words and by using tfidf we assign some weights to words
# we also need to format the words in the text (punctuations, links, cappitals, etc)
# in conclusion in this next part we convert text in machine readable format
```

```python
# to convert text into readable form we import regex
import re
```

```python
def wordopt(text):

  # converting to lower case
  text = text.lower()

  # removing URL's
  text = re.sub (r'https?://\S+|www\.\S+','',text)

  # Remove HTML tags
  text = re.sub(r'<.*?>', '', text)

  # Remove punctuation
  text = re.sub(r'[^\w\s]', '', text)

  # Remove digits
  text = re.sub(r'\d', '', text)

  # Remove newline characters
  text = re.sub(r'\n', '', text)

  return text
```

```python
# applying the wordopt function
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
# defining independant variable
x = news['text']
```

```python
# defining dependant variable
y = news['label']
```

```python
x
```

```python
y
```

```python
# now we divide dataset into training and test data
from sklearn.model_selection import train_test_split
```

```python
# dividing training and test data on the basis of 30%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# converting text in x_train data into numeric data (vectors)
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# creating an instance
vectorization = TfidfVectorizer()
```

```python
# creating vectors for training data
xv_train = vectorization.fit_transform(x_train)
```

```python
# creating vectors for testing data
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
# Applying Models
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# creating instance LR for logisticRegression
LR = LogisticRegression()
```

```python
# training model with the data
LR.fit(xv_train, y_train)
```

```python
# checking the prediction
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```
